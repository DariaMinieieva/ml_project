```python
with open("/kaggle/working/config.yaml", "w") as f:
    f.write('''# Dataset paths relative to the yolov5 folder
train: ../input/without-games/tank_without_games_new/train
val:   ../input/without-games/tank_without_games_new/valid
test:  ../input/without-games/tank_without_games_new/test

# Number of classes
nc: 1

# Class names 0 - cat, 1 - dog
names: ['tank']''')
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(data="/kaggle/working/config.yaml", epochs=50)
```

```python
with open("/kaggle/working/val_config.yaml", "w") as f:
    f.write('''# Dataset paths relative to the yolov5 folder
train:
val:  ../input/without-games/tank_without_games_new/valid

# Number of classes
nc: 1

# Class names 0 - cat, 1 - dog
names: ['tank']''')
    
best_model_val = YOLO("/kaggle/input/no-games-best-model/best.pt")
best_model_val.val(data="/kaggle/working/val_config.yaml")
```

```python
with open("/kaggle/working/test_config.yaml", "w") as f:
    f.write('''# Dataset paths relative to the yolov5 folder
train:
val:  ../input/without-games/tank_without_games_new/test

# Number of classes
nc: 1

# Class names 0 - cat, 1 - dog
names: ['tank']''')
    
best_model_test = YOLO("/kaggle/input/no-games-best-model/best.pt")
best_model_test.val(data="/kaggle/working/test_config.yaml")
```
